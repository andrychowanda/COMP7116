```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img = cv2.imread('Jubilee.jpg')
plt.imshow(img)
plt.show()
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
img_result = cv2.cornerHarris(gray, 2, 3, 0.04)
img_result = cv2.dilate(img_result, None)
plt.imshow(img_result, cmap='gray')
plt.show()
```

```python
th = 0.2 * img_result.max()
img_result_corner = img
for i in range(0, img_result.shape[0]):
    for j in range(0, img_result.shape[1]):
        if(img_result[i,j] > th):
            cv2.circle(img_result_corner, (j, i), 1, (255,0,0), 1)
plt.imshow(img_result_corner)
plt.show()
```
