```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Filtering - AVG
img = cv2.imread("Lena.jpg") 
img_filter = np.ones((7, 7), np.float32) / 49
img_result = cv2.filter2D(img, -1, img_filter)
cv2.imwrite("Lena_Smooth.jpg", img_result)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
img_result = cv2.cvtColor(img_result, cv2.COLOR_BGR2RGB)
plt.imshow(img_result)
plt.show()
```

```python
#Filtering - Gaussian
img_result = cv2.GaussianBlur(img,(7,7),0)
cv2.imwrite("Lena_Gaussian.jpg", img_result)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
img_result = cv2.cvtColor(img_result, cv2.COLOR_BGR2RGB)
plt.imshow(img_result)
plt.show()
```

```python
#Filtering - Median
img_result = cv2.medianBlur(img, 7)
cv2.imwrite("Lena_Median.jpg", img_result)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
img_result = cv2.cvtColor(img_result, cv2.COLOR_BGR2RGB)
plt.imshow(img_result)
plt.show()
```

```python
#Filtering - Canny
img = cv2.imread("Lena.jpg", 0) 
img_result = cv2.Canny(img, 80, 120)
cv2.imwrite("Lena_Canny.jpg", img_result)

plt.imshow(img, cmap="gray")
plt.show()
plt.imshow(img_result, cmap="gray")
plt.show()
```

```python
#Filtering - Laplace
img = cv2.imread("Lena.jpg", 0) 
img_result = cv2.Laplacian(img, -1)
cv2.imwrite("Lena_Laplacian.jpg", img_result)

plt.imshow(img, cmap="gray")
plt.show()
plt.imshow(img_result, cmap="gray")
plt.show()
```

```python
#Filtering - Laplace
img = cv2.imread("Lena.jpg", 0) 
img_result = cv2.Sobel(img, -1, 1, 1, ksize = 5)
cv2.imwrite("Lena_Sobel.jpg", img_result)

plt.imshow(img, cmap="gray")
plt.show()
plt.imshow(img_result, cmap="gray")
plt.show()
```

```python
#Convolution
from scipy import signal, misc

img = cv2.imread("Jubilee.jpg", 0) 

img_filter1 = np.array([[ 10, 0,  -10],
                       [ 10, 0, -10],
                       [ 10, 0,  -10]])

img_filter2 = np.array([[ 10, 10,  10],
                       [ 0, 0, 0],
                       [ -10, -10, -10]])

img_result = signal.convolve2d(img, img_filter1, boundary='fill', mode='same')
img_result2 = signal.convolve2d(img, img_filter2, boundary='fill', mode='same')
plt.imshow(img, cmap="gray")
plt.show()
plt.imshow(np.absolute(img_result), cmap="gray")
plt.show()
plt.imshow(np.absolute(img_result2), cmap="gray")
plt.show()
```
