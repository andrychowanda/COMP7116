```python
from skimage.feature import hog
import numpy as np
import cv2
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (16,9)
```

```python
img = cv2.imread('Jubilee.jpg')
plt.imshow(img)
plt.show()
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
hog_features, hog_image = hog(img, visualize = True, block_norm = "L2-Hys", pixels_per_cell=(32, 32))
plt.imshow(hog_image, cmap = "Greys")
plt.show()
hog_features, hog_image = hog(img, visualize = True, block_norm = "L2-Hys", pixels_per_cell=(8, 8))
plt.imshow(hog_image, cmap = "Greys")
plt.show()
```
