```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Negetive
img = cv2.imread('lung.jpg',0)  
m,n = img.shape 
L = img.max() 
img_neg = L - img
cv2.imwrite('lung_gray.png', img) 
cv2.imwrite('lung_neg.png', img_neg) 

plt.imshow(img, cmap="gray")
plt.show()
plt.imshow(img_neg, cmap="gray")
plt.show()
```

```python
#Thresholding
a = 200
b = 230
img_neg_t = img_neg
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        if a < img[i, j] < b:
            img_neg_t[i, j] = L
        else:
            img_neg_t[i, j] = img[i, j] 
cv2.imwrite('lung_neg_t.png', img_neg_t)

plt.imshow(img_neg_t, cmap="gray")
plt.show()
```
